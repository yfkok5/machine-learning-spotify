```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in data
data = Path("Resources/spotify_music_target.csv")
df = pd.read_csv(data)
df.head()
```

```python
df.columns
```

```python
raw_data = df.drop(['Unnamed: 0',"artist","track","type","id","uri","track_href","analysis_url","time_signature","key","duration_ms","liveness"],axis=1)
raw_data.head(49)
```

```python
# Standardize data
spotify_scaled = StandardScaler().fit_transform(raw_data)
print(spotify_scaled[0:5])
```

```python
spotify = np.array(spotify_scaled)
```

```python
#Apply PCA to reduce dimensions from 4 to 2 and create a DataFrame with the principal components data.

# Initialize PCA model
pca = PCA(n_components=5)

# Get two principal components for the data.
spotify_pca = pca.fit_transform(spotify_scaled)
```

```python
# Transform PCA data to a DataFrame
df_spotify_pca = pd.DataFrame(
    data=spotify_pca, columns=["principal component 1", "principal component 2","principal component 3","principal component 4","principal component 5"]
)
df_spotify_pca.head()
```

```python
# Fetch the explained variance
variance_ratio = pca.explained_variance_ratio_.tolist()

variance_ratio[1]
```

```python
print(f"principal component 1: {round(variance_ratio[0]*100,2)}%")
print(f"principal component 2: {round(variance_ratio[1]*100,2)}%")
print(f"principal component 3: {round(variance_ratio[2]*100,2)}%")
print(f"principal component 4: {round(variance_ratio[3]*100,2)}%")
print(f"principal component 5: {round(variance_ratio[4]*100,2)}%")

print(f"We have {round(sum(variance_ratio)*100,2)}% of the information in the original dataset.")
```

```python
# Initialize the K-Means model
model = KMeans(n_clusters=2, random_state=42)

# Fit the model
model.fit(df_spotify_pca)

# Predict clusters
predictions = model.predict(df_spotify_pca)

# Add the predicted class columns
df_spotify_pca["class"] = model.labels_
df_spotify_pca.head(50)
```

```python
df_spotify_pca.hvplot.scatter(
    x="principal component 1",
    y="principal component 2",
    hover_cols=["class"],
    by="class",
)
```
