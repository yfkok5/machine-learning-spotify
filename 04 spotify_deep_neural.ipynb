```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in data
data = Path("Resources/spotify_raw_data.csv")
df = pd.read_csv(data)
df.head(10)
```

```python
#identify column headings
df.columns
```

```python
#drop unnecessary columns
raw_data = df.drop(['Unnamed: 0',"artist","track","type","id","uri","track_href","analysis_url","time_signature"],axis=1)
raw_data.head(20)
```

```python
# Define the features X set and the target y vector
X = raw_data.drop("target",axis=1)
y = raw_data.loc[:, ["target"]]

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=5)

# Scale the data of the features set using the StandardScaler
X_scaler = StandardScaler()
X_scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Apply One-hot encoding to the target labels
enc = OneHotEncoder()
enc.fit(y_train)

encoded_y_train = enc.transform(y_train).toarray()
encoded_y_test = enc.transform(y_test).toarray()
encoded_y_train[0]
```

```python
# Define the model - shallow neural network
model = Sequential()

# Add the first layer where the input dimensions 11 columns of the training data
model.add(Dense(100, activation='relu', input_dim=X_train_scaled.shape[1]))
```

```python
# The output layer has either 0 or 1
y_train.target.value_counts()
number_outputs = 2
```

```python
model.add(Dense(number_outputs, activation="softmax"))

# Compile the model using categorical_crossentropy for the loss function, the adam optimizer,
# and add accuracy to the training metrics
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
# Use the training data to fit (train) the model
# @NOTE: Experiment with the number of training epochs to find the minimum iterations required to achieve a good accuracy
model.fit(
    X_train_scaled,
    encoded_y_train,
    epochs=30,
    shuffle=True,
    verbose=1
)
```

```python
# Evaluate the model using the testing data
model_loss, model_accuracy = model.evaluate(X_test_scaled, encoded_y_test, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Make some predictions with the loaded model
predicted = model.predict(X_test_scaled)
predicted = enc.inverse_transform(predicted).flatten().tolist()

results = pd.DataFrame({
    "Index": y_test.index.values,
    "Actual": y_test.target.values,
    "Predicted": predicted
})

results.head(100)
```

```python
# Print the Classification Report
from sklearn.metrics import classification_report
print(classification_report(results.Actual, results.Predicted))
```
